###### Observations
- The minimum experience of a customer is -3 
which can't be possible
because experience cannot be negative. 
- we will change any negative Experience value by the median
- let's correct the data first then move on
- CCavg minimum value is 0.0 ; suggesting that the customer may not have any credit cards.
- The Minimum value for Mortgage is 0.0 ; this could mean the customers dont own a home.


```python
# loan_data[loan_data['Experience'] < 0]['Experience'].value_counts()

loan_data[loan_data['Experience']<0]['Experience'].value_counts()
```

```python
loan_data[loan_data['Experience'] < 0].count()
```

```python
loan_data['Experience'].replace( to_replace= -1,value = np.nan,inplace = True )
loan_data['Experience'].replace( to_replace= -2,value = np.nan,inplace = True )
loan_data['Experience'].replace( to_replace= -3,value = np.nan,inplace = True )                         
```

```python
# Now we will fill it with median

loan_data['Experience'].fillna(loan_data['Experience'].median(),inplace=True)
```

```python
# Number of people with zero mortgage

loan_data.Mortgage[loan_data['Mortgage']==0].value_counts()
```

```python
# Number of people with zero credit card spending per month

loan_data.CCAvg[loan_data['CCAvg']==0].value_counts()
```
